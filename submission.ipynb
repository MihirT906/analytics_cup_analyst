### [Optional] Option 2: Create a custom episode
Create your own episode by creating a JSON configuration file. This allows you to focus on any specific moment in the game.

**1. Create an episode file**:
```bash
analytics_cup_analyst/src/episodes/custom_episodes/[your_episode_name].json
```

**2. Define your episode parameters**:
```json
{
    "episode_data": {
        "frame_start": <insert starting frame (int)>,
        "frame_end": <insert ending frame (int)>,
        "match_id": <insert match id (int)>
    }
}
```
